# Cosine Similarity

In this section you will construct another similarity metric, now based on the cosinus.

Remember trigonometry (or better, linear algebra!) from your mathematics class? Well this metric is based on trigonometric operations and calculates the angle between two vectors. It might look difficult but it is rather simple. 

## 1. Load the dataset

```python
import pandas as pd
df_books_ratings = pd.read_csv('data/BX-Book-Ratings-Subset.csv', sep=';')
df_books_ratings
```

## 2. Explore the dimensions (shape) of users' rating data

Here are the IDs of two users in our ratings dataset. Collect the ratings made by each user as an array/list. What are their respective ratings' dimensions (shape)? How many books did these users rate respectively?

```python
user_id_a = 277427
user_id_b = 277203

df = df_books_ratings
ratings_a = df[df['User-ID'] == 277427]['Book-Rating']
ratings_b = df[df['User-ID'] == 277203]['Book-Rating']

print(ratings_a.shape)
print(ratings_b.shape)
```

If we are to produce vectors from the users' ratings and apply trigonometric operations on them,  can you see a problem here? Are the vectors of the same dimension? If not, why is this a 'problem'?

```python
# The problem is that these arrays have a different number of dimensions: to make a
# cosine comparison, we need two vectors of the same shape.

# You might also have realised that each element in these vectors corresponds to the
# user's rating for a book, but the nth element in vector a and the nth element in vector
# b are probably not about the same book. So even if the vectors happend to be the same
# shape (or we padded them), doing a pairwise comparison does not make sense.
```

## 3. Vectorize ratings

Can you vectorize the above users' ratings so they have the same dimension? To help you do this, here is sorted  list of all the ISBNs in our dataset. How can you use this list of all the ISBNs to create a (large!) vector for user_id_a?

```python
import numpy as np

ISBNS_array = df['ISBN'].unique()
ISBNS_array = np.sort(ISBNS_array).tolist()

# Select a user

df_user_a = df[df['User-ID'] == user_id_a]
user_a_ratings_by_isbn = dict(zip(df_user_a['ISBN'], df_user_a['Book-Rating']))

user_a_ratings_vector = [
    user_a_ratings_by_isbn.get(isbn, 0)
    for isbn in ISBNS_array
]

print(user_a_ratings_vector)
```

```python
# We can also use pandas to do this

# create empty df with ISBNS columns
dfv = pd.DataFrame(columns=ISBNS_array)

# concat this empty df with the dict for a user's ratings indexed by the user's ID
dfv = pd.concat([dfv, pd.DataFrame(user_a_ratings_by_isbn, index=[user_id_a])])

# make sure we compute with 0 (ignore the warning)
dfv.fillna(0, inplace=True)

# convert to numpy array
dfv.loc[user_id_a].to_numpy()
```

## 4. Helper functions

Below are two functions that (1) retrieve user ratings from a given dataset and (2) vectorize these ratings according to a certain dimension (all ISBNs).

```python
def get_user_ratings(user_id, df_subset):
    
    df_user = df_subset[df_subset['User-ID'] == user_id]
        
    return dict(zip(df_user['ISBN'], df_user['Book-Rating']))
```

```python
def create_ratings_vector(user_ISBN_rating_dict, all_ISBNS_array):
    
    user_ISBNS = user_ISBN_rating_dict.keys()
    
    return [0 if v not in user_ISBNS else user_ISBN_rating_dict[v] for v in all_ISBNS_array]    
```

## 5. Cosine distance function

Can you finish the writing of the function below that calculates the angle between two vectors have the same dimension? As you can see, use the numpy `dot` and `norm` operators to translate the given formula into code.

```python
from numpy import dot
from numpy.linalg import norm

def cosine_distance(ratings_vector_user_a, ratings_vector_user_b):
    
#     a . b  -> dot(a, b)
#     -----
#     |a||b| -> norm(a) * norm(b)
    
    return dot(ratings_vector_user_a, ratings_vector_user_b) / (norm(ratings_vector_user_a) * norm(ratings_vector_user_b))
    
```

## 6. Calculate distances

Here is the ID of a user in our dataset (you can of course choose another one!).

Can you calculate this user's cosine distance from all the other users in the dataset?

```python
a_user_id = 277427

all_user_ids = df['User-ID'].unique().tolist()

user_a_ISBN_ratings = get_user_ratings(a_user_id, df)
user_a_ratings_vector = create_ratings_vector(user_a_ISBN_ratings, ISBNS_array)

distances = dict()

for user_b in all_user_ids:
    
    if user_b == a_user_id:
        continue
    
    user_b_ISBN_ratings = get_user_ratings(user_b, df)    
    user_b_rating_vector = create_ratings_vector(user_b_ISBN_ratings, ISBNS_array)
    
    d = cosine_distance(user_a_ratings_vector, user_b_rating_vector)
    distances[user_b] = d

print(distances)
```

Considering the code above, can you make a function that will take as input a given user's ID and calculate its distance from all other users in our dataset?

```python
all_user_ids = df['User-ID'].unique().tolist()

def user_distances(user_id):
    user_a_ISBN_ratings = get_user_ratings(user_id, df)
    user_a_ratings_vector = create_ratings_vector(user_a_ISBN_ratings, ISBNS_array)

    distances = dict()

    for user_b in all_user_ids:

        if user_b == user_id:
            continue
        
        user_b_ISBN_ratings = get_user_ratings(user_b, df)    
        user_b_rating_vector = create_ratings_vector(user_b_ISBN_ratings, ISBNS_array)
        
        d = cosine_distance(user_a_ratings_vector, user_b_rating_vector)
        distances[user_b] = d

    return distances

    
```
